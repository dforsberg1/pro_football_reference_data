```python
def clean_html_string(html_string):
    # Remove all tags and keep only text content
    html_string = re.sub('<[^>]+>', '', html_string)
    # Remove extra whitespace
    html_string = re.sub('\s+', ' ', html_string).strip()
    return html_string
```

```python
# function to split text by month name
def split_text_by_month(text):
    # find positions of all month names in the text
    month_positions = [(m.start(), m.group()) for m in re.finditer(r'(January|February|March|April|May|June|July|August|September|October|November|December)\s+\d+', text)]
    # split the text at the positions before the month names
    split_texts = []
    start = 0
    for pos in month_positions:
        split_texts.append(text[start:pos[0]].strip())
        start = pos[0]
    split_texts.append(text[start:].strip())
    return split_texts
```

```python
def web_scraper(code_number):
    # Replace all occurrences of _000 with the given code_number
    import inspect
    current_script = inspect.getsource(web_scraper)
    code_str = "".join(current_script)
    code_str = code_str.replace("_000", "_" + str(code_number))

    # Create an empty list to store all the rows
    all_rows = []

    # Loop through all the URLs in the player_urls dataframe
    player_urls_name = f"player_urls_{code_number}"
    for url in eval(player_urls_name)['urls']:
        # Send a GET request to the URL and retrieve the HTML source code
        response = requests.get(url)
        html_source = response.text

        # Split the HTML source code into lines
        lines = html_source.split("\n")

        # Loop through the lines and extract the desired lines
        desired_lines = []
        for line in lines:
            if line.startswith('	    <ul class="transactions news_stories">') or line.startswith('     <ul class="transactions news_stories more">'):
                desired_lines.append(line)

        # Create a list of dictionaries, each representing a row of data
        rows = []
        for line in desired_lines:
            row = {'url': url, 'html': clean_html_string(line)}
            rows.append(row)

        # Append the rows to the all_rows list
        all_rows.extend(rows)
    
    # Should be approx around 50 for each...
    print(f"html lines per 25 players {code_number}: {len(all_rows)}")

    # Create a dataframe from the rows
    df = pd.DataFrame(all_rows)

    # Split the 'html' column by month and explode the resulting list into separate rows
    df = df.assign(html=df['html'].apply(split_text_by_month)).explode('html').reset_index(drop=True)

    # Drop empty rows
    df = df.dropna()

    # Remove periods and extra whitespace from every row
    df['html'] = df['html'].str.replace('.', '', regex=False).str.replace('  ', '')

    # Remove leading/trailing whitespace and replace empty strings with NaN
    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x).replace('', pd.NA)

    # Drop empty rows again
    df = df.dropna()

    # Define pattern to match
    pattern = re.compile(r'\sShow [1-9][0-9]* more transactions')

    # Replace pattern with an empty string
    df['html'] = df['html'].str.replace(pattern, '')

    # Set the maximum column width to 1000
    pd.set_option('display.max_colwidth', 1000)

    file_name = f"player_transactions_{code_number}"

    df.to_excel(file_name + ".xlsx", index=False)

    # Return the dataframe
    # return df
```
